```
#@title Daily Elevator Niche Post Generator
!pip install openai --quiet
import openai, datetime, json, requests, os

openai.api_key = "sk-XXX"  # Use OpenRouter key (free)

pillars = ["Diagnostics","Energy","Parts","Compliance","Case","Myth"]
today = datetime.date.today().toordinal()
pillar = pillars[today % 6]

prompt = f"""
You are a Ugandan lift engineer. Write a 120-word micro-lesson for facility managers on how to {pillar}-related reduce downtime or energy. Include a single actionable tip and end with a CTA to comment.
Keep tone conversational. No emojis.
"""

res = openai.ChatCompletion.create(
  model="gpt-3.5-turbo",
  messages=[{"role":"user","content":prompt}]
)
text = res["choices"][0]["message"]["content"]

# Save to Google Drive
with open("/content/today_post.txt","w") as f:
  f.write(text)
```
